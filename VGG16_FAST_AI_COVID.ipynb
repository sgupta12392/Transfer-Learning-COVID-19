## Using a Pre-trained Model trained on ImageNet Dataset for Transfer Learning 

```python
vgg16model=models.vgg16_bn(pretrained=True)
vgg16model
```

```python
num_inputs = vgg16model.classifier[6].in_features
print(num_inputs)     # number of inputs coming to the last output layer of the RESNET model. we are discrading last ouput layer of resnet and adding more layers for our model
```

### Removing the latter layers of the pre-trained model :

```python
vgg16model.classifier= nn.Sequential(*[vgg16model.classifier[i] for i in range(4)])
print(vgg16model.classifier)
```

```python
vgg16model
```

### Adding a customized head with additional layer for our model :

```python
n_classes=3
head= nn.Sequential(
    #AdaptiveConcatPool2d((4,4)),
    #Flatten(),
    vgg16model,
    nn.ReLU(),
    nn.Dropout(0.4),
    nn.Linear(num_inputs, 256), 
    nn.ReLU(), 
    nn.BatchNorm1d(256),
    nn.Dropout(0.4),
    nn.Linear(256, n_classes), 
    #nn.LogSoftmax(dim=1)
    )
```

```python
head
```

```python
learn=Learner(X,head, loss_func=nn.CrossEntropyLoss(), metrics=[accuracy])
```

```python
print(learn.summary)
```

```python
print(learn.summary())    
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.freeze_to(-1)
```

```python
learn.fit_one_cycle(20, max_lr=1e-02)
```

```python
learn.recorder.plot_losses()
```

```python
learn.unfreeze()
```

```python
learn.fit_one_cycle(20, max_lr=1e-02)
```

```python
learn.show_results(ds_type=DatasetType.Valid)
```

```python
#### between train vs validation data 
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```
